```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('/kaggle/input/weather-dataset/weather_data.csv')
```

```python
# Convert timestamp to datetime format
df['_time'] = pd.to_datetime(df['_time'])
```

```python
plt.figure(figsize=(24, 11))

# Temperature line chart
plt.plot(df['_time'], df['temperature'], linestyle='-', linewidth=2, color='blue')
plt.title('Temperature Over Time', fontsize=22)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Temperature (Â°C)', fontsize=18)

# Formatting timestamps
date_format = mdates.DateFormatter('%d.%m.%Y %H:%M')
plt.gca().xaxis.set_major_formatter(date_format)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.75, color='gray')
plt.gca().set_facecolor('#f9f9f9')
plt.savefig('temperature_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(24, 11))

# Feels Like Temperature line chart
plt.plot(df['_time'], df['feels_like'], linestyle='-', linewidth=2, color='red')
plt.title('Feels Like Temperature Over Time', fontsize=22)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Feels Like Temperature (Â°C)', fontsize=18)

# Formatting timestamps
date_format = mdates.DateFormatter('%d.%m.%Y %H:%M')
plt.gca().xaxis.set_major_formatter(date_format)

plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.75, color='gray')
plt.gca().set_facecolor('#f9f9f9')

# GrafiÄi kaydetme
plt.savefig('feels_like_temperature_over_time.png', dpi=300, bbox_inches='tight')

# Show chart
plt.show()
```

```python
plt.figure(figsize=(24, 11))

# Humidity line chart
plt.plot(df['_time'], df['humidity'], linestyle='-', linewidth=2, color='green')
plt.title('Humidity Over Time', fontsize=22)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Humidity (%)', fontsize=18)

# Formatting timestamps
date_format = mdates.DateFormatter('%d.%m.%Y %H:%M')
plt.gca().xaxis.set_major_formatter(date_format)

plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.75, color='gray')
plt.gca().set_facecolor('#f9f9f9')

# GrafiÄi kaydetme
plt.savefig('humidity_over_time.png', dpi=300, bbox_inches='tight')

# Show chart
plt.show()

```

```python
plt.figure(figsize=(24, 11))

# Wind Speed ââline chart
plt.plot(df['_time'], df['wind_speed'], linestyle='-', linewidth=2, color='blue')
plt.title('Wind Speed Over Time', fontsize=22)
plt.xlabel('Time', fontsize=18)
plt.ylabel('Wind Speed (km/h)', fontsize=18)

# Formatting timestamps
date_format = mdates.DateFormatter('%d.%m.%Y %H:%M')
plt.gca().xaxis.set_major_formatter(date_format)
plt.xticks(rotation=45)
plt.grid(True, which='both', linestyle='--', linewidth=0.75, color='gray')
plt.gca().set_facecolor('#f9f9f9')
plt.savefig('wind_speed_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(df['_time'], df['temperature'], label='Temperature')
plt.plot(df['_time'], df['humidity'], label='Humidity')
plt.plot(df['_time'], df['wind_speed'], label='Wind Speed')
plt.plot(df['_time'], df['feels_like'], label='Feels Like Temperature')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Trends Over Time')
plt.legend()
plt.grid(True)
plt.savefig('trends_over_time.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['temperature'], model='additive', period=24)
result.plot()
plt.savefig('temperature.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
heatmap_data = pd.pivot_table(df, values='temperature', index=df['_time'].dt.hour, columns=df['_time'].dt.date)
sns.heatmap(heatmap_data, cmap="YlGnBu")
plt.title('Temperature Heatmap')
plt.xlabel('Date')
plt.ylabel('Hour of the Day')
plt.savefig('temperature_heatmap.png', dpi=300, bbox_inches='tight')
plt.show()
```
